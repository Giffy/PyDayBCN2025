In this demo, we move beyond fixing errors to **optimizing solutions**. We'll use evolutionary principles:

```

```

## The Problem: Complex Password Generation

We want to generate passwords that meet **all** of these criteria:

| Rule | Points | Description |
|------|--------|-------------|
| Length | 20 pts | Between 12-20 characters |
| Digits | 15 pts | At least 2 digits |
| Case Mix | 15 pts | Both uppercase and lowercase |
| Emoji | 20 pts | At least 1 emoji |
| Special | 15 pts | Contains !@#$%^&* |
| Digit Sum | 15 pts | Digits sum to exactly 15 |

**Maximum score: 100 points**

This is hard because the rules interact! Adding digits to reach sum=15 might push you over the length limit.

## The Fitness Function

In evolutionary algorithms, the **fitness function** determines how "good" each candidate is. It's the selection pressure that drives evolution.

Our fitness function scores each password from 0-100 based on how many rules it satisfies.

````python
import re
def evaluate_password(password: str) -> dict:
    """
    Fitness function: Score a password based on complex rules.

    Args:
        password: The password string to evaluate

    Returns:
        Dict with 'password', 'score' (0-100), and 'criteria' breakdown
    """
    score = 0
    criteria = {}

    # Rule 1: Length between 12-20 chars (20 points)
    if 12 <= len(password) <= 20:
        criteria["length"] = 20
        score += 20
    else:
        criteria["length"] = 0

    # Rule 2: Contains at least 2 digits (15 points)
    digits = sum(c.isdigit() for c in password)
    if digits >= 2:
        criteria["digits"] = 15
        score += 15
    else:
        criteria["digits"] = 0

    # Rule 3: Contains uppercase and lowercase (15 points)
    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    if has_upper and has_lower:
        criteria["case_mix"] = 15
        score += 15
    else:
        criteria["case_mix"] = 0

    # Rule 4: Contains at least 1 emoji (20 points)
    emoji_pattern = re.compile("[\U0001F300-\U0001F9FF]")
    if emoji_pattern.search(password):
        criteria["emoji"] = 20
        score += 20
    else:
        criteria["emoji"] = 0

    # Rule 5: Contains special characters !@#$%^&* (15 points)
    special = set("!@#$%^&*")
    if any(c in special for c in password):
        criteria["special"] = 15
        score += 15
    else:
        criteria["special"] = 0

    # Rule 6: Digits sum to exactly 15 (15 points)
    digit_sum = sum(int(c) for c in password if c.isdigit())
    if digit_sum == 15:
        criteria["digit_sum_15"] = 15
        score += 15
    else:
        criteria["digit_sum_15"] = 0

    return {"password": password, "score": score, "criteria": criteria, "digit_sum": digit_sum}


# Test the fitness function
print("üß™ Testing fitness function:\n")
test_passwords = [
    "simple",           # Too short, no criteria met
    "Password123!",     # Good but no emoji, digits sum wrong
    "Pass96!üîê",        # Has emoji! Let's see...
]

for pw in test_passwords:
    result = evaluate_password(pw)
    print(f"Password: {result['password']}")
    print(f"Score: {result['score']}/100 (digit sum: {result['digit_sum']})")
    print(f"Criteria: {result['criteria']}\n")
````

## Generation 0: Initial Population

Every evolutionary algorithm needs an initial population. We'll ask the LLM to generate 5 initial password candidates.

Note: This first generation is essentially random exploration. The LLM will try, but probably won't hit 100/100 on the first try.

````python
def generate_initial_population(n: int = 5) -> list:
    """
    Generate n password candidates using LLM.

    Args:
        n: Number of passwords to generate

    Returns:
        List of password strings
    """
    prompt = f"""Generate {n} creative password candidates that try to meet these criteria:
    - Length between 12-20 characters
    - At least 2 digits
    - Mix of uppercase and lowercase letters
    - At least 1 emoji (like üîê, üöÄ, üí™, etc.)
    - Special characters (!@#$%^&*)
    - Digits should sum to exactly 15 (e.g., 9+6=15, or 8+7=15, or 5+5+5=15)

    Return ONLY the {n} passwords, one per line, nothing else."""


    response = client.chat(
        model='qwen2.5:0.5b',
        messages=[{'role': 'user', 'content': prompt}],
        options={ 'temperature': 0.2 },
        )

    # response = client.chat.completions.create(
    #     model="gpt-4o-mini",
    #     messages=[{"role": "user", "content": prompt}],
    #     temperature=0.9  # High temperature for diversity
    # )

    #passwords = response.choices[0].message.content.strip().split("\n")
    passwords = response.message.content.strip().split("\n")
    # Clean up and limit to n passwords
    return [p.strip() for p in passwords if p.strip()][:n]
````

````python
# Generate and evaluate initial population
print("üß¨ Generation 0: Initial Population")
print("=" * 50)
print()

population = generate_initial_population(5)
print(population)
evaluated = [evaluate_password(p) for p in population]

# Sort by score (best first)
evaluated.sort(key=lambda x: x["score"], reverse=True)

for i, item in enumerate(evaluated):
    print(f"{i+1}. Score: {item['score']}/100 - {item['password']}")
    print(f"   Digit sum: {item['digit_sum']} (target: 15)")
    print(f"   Criteria: {item['criteria']}")
    print()
````

## Selection & Mutation via LLM

This is where the magic happens! Traditional genetic algorithms use:
- **Selection**: Keep the fittest individuals
- **Crossover**: Combine traits from two parents
- **Mutation**: Random changes to explore new solutions

We'll use the **LLM as an intelligent mutation/crossover operator**:
- It sees the best parents and their scores
- It understands which criteria are missing
- It creates new candidates that combine the best traits

````python
def evolve_population(parents: list, generation: int) -> list:
    """
    Take the 2 best parents and ask LLM to create mutations/crossovers.

    Args:
        parents: List of evaluated password dicts (sorted by score)
        generation: Current generation number (for logging)

    Returns:
        List of new password strings
    """
    parent1 = parents[0]["password"]
    parent2 = parents[1]["password"]
    p1_score = parents[0]["score"]
    p2_score = parents[1]["score"]
    p1_criteria = parents[0]["criteria"]
    p2_criteria = parents[1]["criteria"]
    p1_digit_sum = parents[0]["digit_sum"]
    p2_digit_sum = parents[1]["digit_sum"]

    prompt = f"""You are evolving passwords through a genetic algorithm. Generation {generation}.

    PARENT 1 (Score: {p1_score}/100): {parent1}
    - Criteria met: {p1_criteria}
    - Current digit sum: {p1_digit_sum} (target: 15)

    PARENT 2 (Score: {p2_score}/100): {parent2}
    - Criteria met: {p2_criteria}
    - Current digit sum: {p2_digit_sum} (target: 15)

    TARGET CRITERIA (to maximize score to 100):
    - Length 12-20 chars (20 pts)
    - At least 2 digits (15 pts)
    - Mix upper/lowercase (15 pts)
    - At least 1 emoji (20 pts)
    - Special chars !@#$%^&* (15 pts)
    - Digits sum to EXACTLY 15 (15 pts) - e.g., 9+6=15, 8+7=15, 5+5+5=15

    Create 5 NEW passwords by combining and mutating the best traits from both parents.
    Focus especially on:
    1. Getting the digit sum to exactly 15
    2. Improving any criteria that parents are missing
    3. Keeping the traits that are already working

    Return ONLY 5 passwords, one per line, nothing else."""

    response = client.chat(
        model='qwen2.5:0.5b',
        messages=[{'role': 'user', 'content': prompt}],
        options={ 'temperature': 0.8 }, # Slightly lower for more focused improvement
        )
    children = response.message.content.strip().split("\n")

    # response = client.chat.completions.create(
    #     model="gpt-4o-mini",
    #     messages=[{"role": "user", "content": prompt}],
    #     temperature=0.8  # Slightly lower for more focused improvement
    # )
    #children = response.choices[0].message.content.strip().split("\n")

    return [c.strip() for c in children if c.strip()][:5]
````

## The Evolution Loop

Now let's run evolution for 3 generations and watch the population improve!

Each generation:
1. Select the 2 best parents
2. Ask LLM to create 5 children through crossover/mutation
3. Evaluate and rank the new generation
4. Repeat

````python
NUM_GENERATIONS = 10

# Track best scores across generations for visualization
best_scores = [evaluated[0]["score"]]

for gen in range(1, NUM_GENERATIONS + 1):
    print(f"\n{'='*60}")
    print(f"üß¨ Generation {gen}: Evolution")
    print(f"{'='*60}\n")

    # Select top 2 parents
    top_parents = evaluated[:2]
    print(f"üë®‚Äçüë©‚Äçüëß‚Äçüë¶ Parents selected:")
    print(f"  1. {top_parents[0]['password']}")
    print(f"     Score: {top_parents[0]['score']}/100, Digit sum: {top_parents[0]['digit_sum']}")
    print(f"  2. {top_parents[1]['password']}")
    print(f"     Score: {top_parents[1]['score']}/100, Digit sum: {top_parents[1]['digit_sum']}")
    print()

    # Generate children through crossover/mutation
    print("üîÑ Generating offspring via LLM crossover/mutation...\n")
    children = evolve_population(top_parents, gen)

    # Evaluate new generation
    evaluated = [evaluate_password(p) for p in children]
    evaluated.sort(key=lambda x: x["score"], reverse=True)

    # Track best score
    best_scores.append(evaluated[0]["score"])

    print("üå± New Generation:")
    for i, item in enumerate(evaluated):
        status = "üèÜ" if item["score"] == 100 else "  "
        print(f"{status} {i+1}. Score: {item['score']}/100 - {item['password']}")
        print(f"      Digit sum: {item['digit_sum']} (target: 15)")

    # Check if we found a perfect solution
    if evaluated[0]["score"] == 100:
        print("\nüéâ PERFECT SCORE ACHIEVED!")
        break
````

Let's see the best password we evolved and how scores improved across generations.

````python
print(f"{'='*60}")
print(f"üèÜ EVOLUTION COMPLETE")
print(f"{'='*60}")
print()

best = evaluated[0]
print(f"Best Password: {best['password']}")
print(f"Final Score: {best['score']}/100")
print(f"Digit Sum: {best['digit_sum']} (target: 15)")
print()
print("Criteria Breakdown:")
for criterion, points in best['criteria'].items():
    status = "‚úÖ" if points > 0 else "‚ùå"
    print(f"  {status} {criterion}: {points} pts")

print()
print("üìà Score progression across generations:")
for i, score in enumerate(best_scores):
    bar = "‚ñà" * (score // 5)
    print(f"  Gen {i}: {bar} {score}/100")
````

## Part 2: Evolving Code (FizzBuzz)

Now let's apply the same evolutionary principles to **actual Python code**!

Instead of evolving strings, we'll evolve FizzBuzz implementations:
- Return "Fizz" for multiples of 3
- Return "Buzz" for multiples of 5  
- Return "FizzBuzz" for multiples of both
- Return the number otherwise

The LLM will generate different implementations and we'll evolve them for:
- **Correctness**: Does it produce the right output?
- **Brevity**: Is the code concise?

This demonstrates that LLMs understand code **semantics**, not just syntax!

````python
def evaluate_fizzbuzz(code: str) -> dict:
    """
    Fitness function for FizzBuzz implementations.

    Scores based on:
    - Correctness (70 pts): Output matches expected for n=1..20
    - Brevity (15 pts): Shorter code is better
    - Pythonic (15 pts): Base bonus for valid code
    """
    # Expected output for n=1..20
    expected = []
    for i in range(1, 21):
        if i % 15 == 0: expected.append("FizzBuzz")
        elif i % 3 == 0: expected.append("Fizz")
        elif i % 5 == 0: expected.append("Buzz")
        else: expected.append(str(i))

    # Execute the code
    try:
        namespace = {}
        exec(code, namespace)
        if 'fizzbuzz' not in namespace:
            return {"code": code, "score": 0, "error": "No fizzbuzz function", "correct": 0}

        # Test correctness
        results = [namespace['fizzbuzz'](i) for i in range(1, 21)]
        correct = sum(1 for r, e in zip(results, expected) if str(r) == e)
        correctness_score = int(70 * correct / 20)

        # Brevity score (fewer chars = better, max 15 pts)
        char_count = len(code.replace(" ", "").replace("\n", ""))
        brevity_score = max(0, 15 - char_count // 20)

        # Pythonic bonus (base score for working code)
        pythonic_score = 15 if correct > 0 else 0

        total = correctness_score + brevity_score + pythonic_score
        return {
            "code": code,
            "score": total,
            "correct": correct,
            "chars": char_count,
            "correctness": correctness_score,
            "brevity": brevity_score,
            "pythonic": pythonic_score
        }
    except Exception as e:
        return {"code": code, "score": 0, "error": str(e), "correct": 0}


# Test with a simple implementation
test_code = '''def fizzbuzz(n):
    if n % 15 == 0:
        return "FizzBuzz"
    elif n % 3 == 0:
        return "Fizz"
    elif n % 5 == 0:
        return "Buzz"
    else:
        return str(n)'''

result = evaluate_fizzbuzz(test_code)
print("üß™ Testing FizzBuzz fitness function:\n")
print(f"Score: {result['score']}/100")
print(f"Correct: {result['correct']}/20")
print(f"Character count: {result['chars']}")
````

````python
def generate_fizzbuzz_population(n: int = 4) -> list:
    """Generate n different FizzBuzz implementations using LLM."""
    prompt = f"""Generate {n} DIFFERENT Python implementations of FizzBuzz.
                Each must be a function called `fizzbuzz(n)` that:
                - Returns "Fizz" if n is divisible by 3
                - Returns "Buzz" if n is divisible by 5
                - Returns "FizzBuzz" if divisible by both
                - Returns str(n) otherwise

                Make each implementation use a DIFFERENT approach:
                1. Traditional if/elif/else
                2. String concatenation approach
                3. Dictionary/mapping approach
                4. One-liner or compact style

                Return ONLY the {n} function definitions, separated by blank lines.
                No explanations, no markdown, no numbering.
                """

    response = client.chat(
        model='qwen2.5:0.5b',
        messages=[{'role': 'user', 'content': prompt}],
        options={ 'temperature': 0.9 },
    )
    content = response.message.content

    # response = client.chat.completions.create(
    #     model="gpt-4o-mini",
    #     messages=[{"role": "user", "content": prompt}],
    #     temperature=0.9
    # )
    # content = response.choices[0].message.content

    # Parse functions from response
    functions = []
    current = []

    for line in content.split('\n'):
        if line.strip().startswith('def fizzbuzz') and current:
            functions.append('\n'.join(current))
            current = []
        if line.strip():  # Skip empty lines at start
            current.append(line)

    if current:
        functions.append('\n'.join(current))

    # Clean and filter valid functions
    result = []
    for f in functions:
        f = f.strip()
        if 'def fizzbuzz' in f:
            result.append(f)

    return result[:n]


# Generate initial population
print("üß¨ FizzBuzz Generation 0: Initial Population")
print("=" * 50)
print()

fb_population = generate_fizzbuzz_population(4)
fb_evaluated = [evaluate_fizzbuzz(code) for code in fb_population]
fb_evaluated.sort(key=lambda x: x["score"], reverse=True)

for i, item in enumerate(fb_evaluated):
    print(f"Implementation {i+1}: Score {item['score']}/100")
    print(f"  Correct: {item['correct']}/20, Chars: {item.get('chars', 'N/A')}")
    print(f"  Code preview: {item['code'][:60]}...")
    print()
````

````python
def evolve_fizzbuzz(parents: list, generation: int) -> list:
    """Evolve FizzBuzz implementations through mutation and crossover."""
    parent1 = parents[0]
    parent2 = parent1
    if len(parents)>1:
        parent2 = parents[1]
        

    prompt = f"""You are evolving FizzBuzz code. Generation {generation}.
                PARENT 1 (Score: {parent1['score']}/100, Correct: {parent1['correct']}/20):
                ```python
                {parent1['code']}
                ```

                PARENT 2 (Score: {parent2['score']}/100, Correct: {parent2['correct']}/20):
                ```python
                {parent2['code']}
                ```

                Create 4 NEW fizzbuzz implementations by:
                1. MUTATION: Modify parent 1 slightly (change logic, simplify, etc.)
                2. MUTATION: Modify parent 2 slightly
                3. CROSSOVER: Combine best parts of both parents
                4. INNOVATION: Try a completely new approach (one-liner, dict-based, etc.)

                Requirements:
                - Each must be a complete `def fizzbuzz(n):` function
                - Must return "Fizz", "Buzz", "FizzBuzz", or str(n)
                - Aim for correctness AND brevity

                Return ONLY the 4 function definitions, separated by blank lines.
                No explanations, no markdown."""


    response = client.chat(
        model='qwen2.5:0.5b',
        messages=[{'role': 'user', 'content': prompt}],
        options={ 'temperature': 0.8 },
    )
    content = response.message.content
    print(content)
    # response = client.chat.completions.create(
    #     model="gpt-4o-mini",
    #     messages=[{"role": "user", "content": prompt}],
    #     temperature=0.8
    # )
    # content = response.choices[0].message.content

    # Parse functions (same logic as initial population)
    functions = []
    current = []

    for line in content.split('\n'):
        if line.strip().startswith('def fizzbuzz') and current:
            functions.append('\n'.join(current))
            current = []
        if line.strip():
            current.append(line)

    if current:
        functions.append('\n'.join(current))

    result = []
    for f in functions:
        f = f.strip()
        if 'def fizzbuzz' in f:
            result.append(f)

    return result[:4]
````

````python
# Run FizzBuzz evolution for 3 generations
FB_GENERATIONS = 3
fb_best_scores = [fb_evaluated[0]['score']]

for gen in range(1, FB_GENERATIONS + 1):
    print(f"\n{'='*60}")
    print(f"üß¨ FizzBuzz Generation {gen}: Evolution")
    print(f"{'='*60}\n")

    # Select top 2 parents
    top_parents = fb_evaluated[:2]
    print(f"üë®‚Äçüë©‚Äçüëß‚Äçüë¶ Parents selected:")
    print(f"  Parent 1: Score {top_parents[0]['score']}/100")
    if len(top_parents)>1:
        print(f"  Parent 2: Score {top_parents[1]['score']}/100")
    print()

    # Generate children through crossover/mutation
    print("üîÑ Evolving code via LLM mutation/crossover...\n")
    children = evolve_fizzbuzz(top_parents, gen)

    # Evaluate new generation
    fb_evaluated = [evaluate_fizzbuzz(code) for code in children]
    fb_evaluated.sort(key=lambda x: x['score'], reverse=True)

    print(fb_evaluated)
    # Track best score
    #b_best_scores.append(fb_evaluated[0]['score'])

    print("üå± New Generation:")
    for i, item in enumerate(fb_evaluated):
        status = "üèÜ" if item['score'] >= 95 else "  "
        print(f"{status} {i+1}. Score: {item['score']}/100 (Correct: {item['correct']}/20)")

    # Check for high score
    if fb_evaluated[0]['score'] >= 95:
        print("\nüéâ HIGH SCORE ACHIEVED!")
        break
````

````python
print(f"{'='*60}")
print(f"üèÜ FIZZBUZZ EVOLUTION COMPLETE")
print(f"{'='*60}")
print()

best_fb = fb_evaluated[0]
print(f"Best Score: {best_fb['score']}/100")
print(f"Correctness: {best_fb['correct']}/20 tests passed")
print(f"Code length: {best_fb.get('chars', 'N/A')} chars")
print()
print("üìù Winning Implementation:")
print("-" * 40)
print(best_fb['code'])
print("-" * 40)
print()
print("üìà Score progression:")
for i, score in enumerate(fb_best_scores):
    bar = "‚ñà" * (score // 5)
    print(f"  Gen {i}: {bar} {score}/100")
````

## Key Takeaways

### What We Learned

1. **LLM as Intelligent Operator**: Unlike random mutations, LLMs understand *semantics*. They can combine the best aspects of two solutions meaningfully.

2. **Fitness-Driven Selection**: The fitness function defines what "good" means. Evolution finds solutions that maximize it, even with complex, interacting constraints.

3. **Code Evolution**: The same evolutionary principles that work for strings also work for **actual code**. The LLM understands program structure, not just characters.

4. **Exploration vs Exploitation**: High temperature = more diversity (exploration). Lower temperature = more focused improvement (exploitation).

### The Power of LLM-Guided Evolution

Traditional genetic algorithms use random bit flips. LLMs provide:
- **Semantic understanding** of what makes solutions better
- **Directed mutations** that preserve working parts
- **Intelligent crossover** that combines abstract concepts
